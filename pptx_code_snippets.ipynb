````python
#To keep only the first and last values for each unique value in the dictionary, you can follow these steps:
# Initial dictionary
my_dict = {
    1: 'CY', 2: 'CY', 3: 'CY',
    4: 'LY', 5: 'LY', 6: 'LY',
    7: 'LLY', 8: 'LLY', 9: 'LLY'
}

# Dictionary to store the first and last occurrences
result_dict = {}

# Dictionary to store the first occurrence index of each value
first_occurrence = {}

# Dictionary to store the last occurrence index of each value
last_occurrence = {}

# Iterate through the dictionary
for key, value in my_dict.items():
    # Record the first occurrence
    if value not in first_occurrence:
        first_occurrence[value] = key
    # Always update the last occurrence
    last_occurrence[value] = key

# Add first and last occurrences to the result dictionary
for value in first_occurrence:
    result_dict[first_occurrence[value]] = value
    result_dict[last_occurrence[value]] = value

# Sort the dictionary by keys to maintain order
result_dict = dict(sorted(result_dict.items()))

print(result_dict)
````

````python

how to remove the first item of the python dictionary
To remove the first item from a Python dictionary, you can use the following methods. Note that Python dictionaries do not maintain any specific order of elements before Python 3.7. From Python 3.7 onwards, dictionaries maintain the insertion order, which makes it possible to remove the "first" item (the item that was inserted first).

Here are a couple of methods to remove the first item from a dictionary in Python 3.7+:

### Method 1: Using `popitem()`

The `popitem()` method removes the last inserted item from the dictionary. To remove the first item, you would have to convert the dictionary to a list of items and then use `popitem()`.

```python
# Initial dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Remove the first item
first_key = next(iter(my_dict))

    In the expression first_key = next(iter(my_dict)), several functions are combined to achieve the goal of getting the first key from the dictionary. Here's a step-by-step breakdown of what each part does:

    1. iter(my_dict)
    iter(): This function returns an iterator object for the given iterable. In this case, the iterable is the dictionary my_dict.
    Iterator Object: An iterator is an object that contains a countable number of values and can be iterated upon. It returns data one element at a time when called with the next() function.
    For a dictionary, the iterator returned by iter(my_dict) will iterate over the dictionary's keys.

    2. next(iter(my_dict))
    next(): This function retrieves the next item from an iterator. If the iterator is exhausted (no more items), it raises a StopIteration exception.
    When you pass iter(my_dict) to next(), it retrieves the first key from the dictionary's iterator.
    Combining Both: first_key = next(iter(my_dict))
    Create an Iterator: iter(my_dict) creates an iterator for the dictionary keys.
    Get the First Key: next(iter(my_dict)) retrieves the first key from this iterator.

first_item = my_dict.pop(first_key)

# Resulting dictionary
print(my_dict)
print("Removed item:", first_item)
```

### Method 2: Using `pop()` with `next(iter())`

This method directly uses `pop()` with the key obtained from `next(iter())` to remove the first item.

```python
# Initial dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Remove the first item
first_key = next(iter(my_dict))
first_item = my_dict.pop(first_key)

# Resulting dictionary
print(my_dict)
print("Removed item:", first_item)
```

### Method 3: Using `del` with `next(iter())`

Another approach is to use the `del` statement with the key obtained from `next(iter())`.

```python
# Initial dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Remove the first item
first_key = next(iter(my_dict))
del my_dict[first_key]

# Resulting dictionary
print(my_dict)
```

All these methods will effectively remove the first item from the dictionary. Choose the one that you find most readable and suitable for your use case.
```
````

The correlation between Microsoft Office PowerPoint (PPT) and XML lies in the underlying file format used by PowerPoint presentations. Let me break it down for you:

1. **PresentationML (PPTX)**:
   - PowerPoint presentations are stored in a file format called **PresentationML** (PPTX). PresentationML is based on **XML** (eXtensible Markup Language).
   - Each PowerPoint presentation is essentially a collection of XML files bundled together.
   - The main XML file is named `presentation.xml`, which contains the structure and content of the presentation.
   - Other related XML files include:
     - `presProps.xml`: Contains presentation-wide properties.
     - `slideMaster.xml`: Describes slide master layouts.
     - `slideLayout.xml`: Specifies slide layout relationships.
     - `theme.xml`: Holds formatting options through a theme.
   - These XML files collectively define the presentation's layout, content, and formatting.

2. **Integration of Data**:
   - XML files play a crucial role in PowerPoint presentations because they allow seamless integration of data from various sources.
   - By utilizing XML files, presenters can dynamically update their slides with real-time information, keeping their audience engaged and informedÂ¹.

3. **Office Open XML (OOXML)**:
   - PresentationML (PPTX) is part of the broader Office Open XML (OOXML) standard.
   - OOXML is also used for Word and Excel documents, as well as Visio files (starting from Office 2013).
   - However, content can be coerced as Office Open XML only in Office Add-ins created for WordÂ³.

````python
# python pptx will ppt display the styler formatted df in proper format as table

import pandas as pd
from pptx import Presentation
from pptx.util import Inches
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor

# Create a sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40],
    'C': ['foo', 'bar', 'baz', 'qux']
})

# Create a PowerPoint presentation
prs = Presentation()

# Add a slide with a title and content layout
slide_layout = prs.slide_layouts[5]  # Choosing a slide layout without title or other placeholders
slide = prs.slides.add_slide(slide_layout)

# Set title (optional)
title = slide.shapes.title
title.text = "DataFrame Table"

# Define the position and size of the table
x, y, cx, cy = Inches(1), Inches(1), Inches(6), Inches(3)
table = slide.shapes.add_table(df.shape[0] + 1, df.shape[1], x, y, cx, cy).table

# Set column names
for col_idx, col_name in enumerate(df.columns):
    table.cell(0, col_idx).text = col_name

# Add the DataFrame content
for row_idx in range(df.shape[0]):
    for col_idx in range(df.shape[1]):
        table.cell(row_idx + 1, col_idx).text = str(df.iloc[row_idx, col_idx])

# Style the table
for row_idx in range(df.shape[0] + 1):  # Include header row
    for col_idx in range(df.shape[1]):
        cell = table.cell(row_idx, col_idx)
        
        # Center align text
        cell.text_frame.paragraphs[0].alignment = PP_ALIGN.CENTER
        
        # Apply alternate row coloring
        if row_idx == 0:
            # Header row styling
            cell.fill.solid()
            cell.fill.fore_color.rgb = RGBColor(0, 0, 128)  # Dark blue background
            cell.text_frame.paragraphs[0].font.color.rgb = RGBColor(255, 255, 255)  # White text
            cell.text_frame.paragraphs[0].font.bold = True
        elif row_idx % 2 == 0:
            cell.fill.solid()
            cell.fill.fore_color.rgb = RGBColor(242, 242, 242)  # Light gray background

# Save the presentation
prs.save('styled_dataframe_table.pptx')
````

````python
# i want to add a total row after each subgroup of Policy_Name.

import pandas as pd
import numpy as np

# Sample data for illustration purposes
data = {
    'Policy_Name': ['Policy1', 'Policy1', 'Policy2', 'Policy2', 'Policy2'],
    'claim_sub_type': ['Type1', 'Type2', 'Type1', 'Type2', 'Type1'],
    'Hospital Type': ['HospitalA', 'HospitalB', 'HospitalA', 'HospitalB', 'HospitalC'],
    'Claims Count': [5, 3, 8, 6, 2],
    'Liability Amount': [100, 150, 200, 250, 300]
}

claims_df = pd.DataFrame(data)

# Custom aggregation
custom_aggr = {
    'Claims Count': 'count',
    'Liability Amount': np.sum
}

# Perform the groupby and aggregation
hospital_nnw_df = claims_df.groupby(['Policy_Name', 'claim_sub_type', 'Hospital Type']).agg(custom_aggr).fillna(0)

# Reset index to make manipulation easier
hospital_nnw_df = hospital_nnw_df.reset_index()

# Calculate totals for each Policy_Name
totals = hospital_nnw_df.groupby('Policy_Name').sum().reset_index()
totals['claim_sub_type'] = 'Total'
totals['Hospital Type'] = ''

# Append totals to the original DataFrame
hospital_nnw_df = pd.concat([hospital_nnw_df, totals], ignore_index=True)

# Sort the DataFrame to ensure totals are at the end of each group
hospital_nnw_df = hospital_nnw_df.sort_values(by=['Policy_Name', 'claim_sub_type', 'Hospital Type'])

# Optionally, reindex to ensure proper order
hospital_nnw_df = hospital_nnw_df.reset_index(drop=True)

# Unstack the DataFrame as per the original instruction
hospital_nnw_df = hospital_nnw_df.set_index(['Policy_Name', 'claim_sub_type', 'Hospital Type']).unstack().fillna(0)

# Print the final DataFrame
display(hospital_nnw_df)
````

pandas- how to keep non-numerical rows while doing column-wise sum on group by in pandas

To retain non-numeric columns while performing a column-wise sum on grouped data in pandas, you can use a combination of `groupby` and `agg` to specify different aggregation functions for numeric and non-numeric columns. Hereâ€™s a step-by-step guide to achieve this:

### Step-by-Step Guide

1. **Identify Numeric and Non-Numeric Columns**: Separate the numeric columns (that will be summed) from the non-numeric columns.

2. **Group By and Aggregate**: Use the `groupby` function followed by `agg` to apply the sum function to numeric columns and an appropriate function (like `first`, `last`, or a custom function) to non-numeric columns.

### Example Code

```python
import pandas as pd

# Sample DataFrame
data = {
    'Policy_Name': ['A', 'A', 'B', 'B'],
    'Value1': [10, 20, 30, 40],
    'Value2': [1.5, 2.5, 3.5, 4.5],
    'Category': ['X', 'X', 'Y', 'Y']
}

df = pd.DataFrame(data)

# Identify numeric and non-numeric columns
numeric_cols = df.select_dtypes(include='number').columns
non_numeric_cols = df.select_dtypes(exclude='number').columns

# Create a dictionary for the aggregation functions
agg_dict = {col: 'sum' for col in numeric_cols}
agg_dict.update({col: 'first' for col in non_numeric_cols})

# Perform groupby and aggregation
result = df.groupby('Policy_Name').agg(agg_dict).reset_index()

print(result)
```

### Explanation

1. **Identify Numeric and Non-Numeric Columns**:
   ```python
   numeric_cols = df.select_dtypes(include='number').columns
   non_numeric_cols = df.select_dtypes(exclude='number').columns
   ```

   This step separates the numeric columns (`Value1`, `Value2`) from the non-numeric columns (`Category`).

2. **Create Aggregation Dictionary**:
   ```python
   agg_dict = {col: 'sum' for col in numeric_cols}
   agg_dict.update({col: 'first' for col in non_numeric_cols})
   ```

   This dictionary maps each column to its corresponding aggregation function. Numeric columns will be summed, and non-numeric columns will retain the first value encountered in each group.

3. **Perform Groupby and Aggregation**:
   ```python
   result = df.groupby('Policy_Name').agg(agg_dict).reset_index()
   ```

   This step groups the DataFrame by `Policy_Name` and applies the aggregation functions specified in the `agg_dict`. The result is then reset to convert the grouped index back into a column.

### Example Output

The resulting DataFrame will look like this:

```
  Policy_Name  Value1  Value2 Category
0           A      30     4.0        X
1           B      70     8.0        Y
```

In this example, the numeric columns (`Value1`, `Value2`) are summed for each `Policy_Name`, and the non-numeric column (`Category`) retains the first value in each group. This method ensures that you keep both numeric and non-numeric data while performing the aggregation.

````python
# pandas how to calculate % value of a column wrt the total of that column
import pandas as pd

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 30, 40]
}
df = pd.DataFrame(data)

# Calculate the total sum of the 'Values' column
total_sum = df['Values'].sum()

# Calculate the percentage of each value with respect to the total sum
df['Percentage'] = (df['Values'] / total_sum) * 100

# Convert the 'Percentage' column to strings and add the '%' symbol
df['Percentage'] = df['Percentage'].apply(lambda x: f'{x:.2f}%')

# Print the DataFrame
print(df)
````

````python
# Pandas `Styler` object allows you to apply styles to your DataFrame to improve the appearance when rendering to various outputs, like HTML. Here are some ways to use `Styler` for alternate row coloring and other beautifications:

### Basic Usage of Pandas Styler

First, let's create a sample DataFrame:

```python
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40],
    'C': ['foo', 'bar', 'baz', 'qux']
})
```

### 1. Alternate Row Coloring

To apply alternate row coloring, use `applymap` or `apply` with custom functions:

```python
def highlight_alternate_rows(row):
    return ['background-color: #f2f2f2' if row.name % 2 == 0 else '' for _ in row]

styled_df = df.style.apply(highlight_alternate_rows, axis=1)
styled_df
```

### 2. Alternate Row Coloring Using `Styler.applymap_index`

You can also use `Styler.applymap_index` for index-specific styling:

```python
styled_df = df.style.applymap(lambda x: 'background-color: #f2f2f2', subset=pd.IndexSlice[::2, :])
styled_df
```

### 3. Highlighting Specific Cells

Highlight cells based on their values:

```python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

styled_df = df.style.apply(highlight_max)
styled_df
```

### 4. Color Mapping Based on Values

Color the cells based on their values using a gradient:

```python
styled_df = df.style.background_gradient(cmap='viridis')
styled_df
```

### 5. Adding Borders and Font Styles

Customize borders and fonts:

```python
styled_df = df.style.set_properties(**{
    'border-color': 'black',
    'border-style': 'solid',
    'border-width': '1px',
    'font-weight': 'bold',
    'color': 'blue'
})
styled_df
```

### 6. Conditional Formatting

Apply conditional formatting to the DataFrame:

```python
def highlight_greater_than_20(s):
    return ['background-color: lightgreen' if v > 20 else '' for v in s]

styled_df = df.style.apply(highlight_greater_than_20, subset=['B'])
styled_df
```

### 7. Format Numeric Values

Format numeric values with specific precision:

```python
styled_df = df.style.format({'A': "{:.2f}", 'B': "{:.1f}"})
styled_df
```

### 8. Bar Chart within Cells

Show bar charts within cells:

```python
styled_df = df.style.bar(subset=['A', 'B'], color='lightblue')
styled_df
```

### 9. Multiple Styles Combined

Combine multiple styles for a comprehensive effect:

```python
def apply_styles(s):
    return ['background-color: #f2f2f2' if s.name % 2 == 0 else '' for _ in s]

styled_df = df.style\
    .apply(apply_styles, axis=1)\
    .highlight_max(color='yellow')\
    .set_properties(**{'border-color': 'black', 'border-style': 'solid', 'border-width': '1px'})\
    .format({'A': "{:.2f}", 'B': "{:.1f}"})
styled_df
```

### 10. Exporting Styled DataFrame to HTML

To save the styled DataFrame as an HTML file:

```python
html = styled_df.render()
with open('styled_df.html', 'w') as f:
    f.write(html)
```

### Conclusion

The `Styler` object in pandas provides a flexible way to style your DataFrames for better visual appeal. You can customize row colors, cell highlights, font properties, and much more to create a visually pleasing DataFrame output. The examples provided cover a variety of common use cases to get you started.
````
