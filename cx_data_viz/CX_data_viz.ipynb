### 2D Line plot

**2D Line Plot**
used for
- Bivariate Analysis-- i.e works on 2 cols.
- used with categorical-numerical data; numerical-numerical data.
    cat data is place on x-axis & numerical data on y-axis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
plt.ylim(45000, 75000)
plt.show()
```

```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
plt.show() # makes the graph to appear on screen(in ide this is not required though). in text editors like pycharm, or in flask plt.show() is must to display the graph.
```

```python
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
# xlim and ylim are used when there are outlier in the graph generally in y-axis
plt.xlim(2015,2020) # useful when there is spike in the value and graph change due to that.
plt.ylim(40000,75000) # will limit the y-aixs between 40K and 75K
```

### 3D surface plot

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx, yy =np.meshgrid(x,y)
z=xx**2+yy**2
# z=np.sin(xx)+np.cos(yy)
```

```python
fig=plt.figure(figsize=(12,6))

ax=plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)

plt.show()
```

```python
## contour plot-- It projects the 3D grph in 2D

fig=plt.figure(figsize=(12,6))

ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')

fig.colorbar(p)
plt.show()
```

```python
## contourf plot-- It projects the 3D grph in 2D but colors are filled.

fig=plt.figure(figsize=(12,6))

ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')

fig.colorbar(p)
plt.show()
```
